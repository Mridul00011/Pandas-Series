```python
import pandas as pd
import numpy as np
```

# Creating Pandas Series from List

String Data

```python
country=["India","USA","Australia","England"]
con=pd.Series(country)
print(con)
```

Integer Data

```python
runs=[13,20,25,60,100]
total=pd.Series(runs)
print(total)
```

Custom Indexing

```python
marks=[67,57,89,100]
subjects=["Maths","English","Science","Hindi"]
result=pd.Series(marks,index=subjects)
print(result)
```

Setting Name

```python
marks=[67,57,89,100]
subjects=["Maths","English","Science","Hindi"]
scored=pd.Series(marks,index=subjects,name="Mridul ke marks")
print(scored)
```

# Creating Pandas Series from Dictionary

```python
marks={
    "Maths":67,
    "English":57,
    "Science":89,
    "Hindi":100
}
Result=pd.Series(marks,name="Mridul's Marks")
print(Result)
```
